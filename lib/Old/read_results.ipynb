```python
import json
import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
import lib.Metrics as Metrics
from lib.HHS_data import *

num = 1
year = 2015

file_path = os.path.join('results', '209', 'save_data_{}-{}.json'.format(num, year))

with open(file_path, 'r') as json_file:
    data = json.load(json_file)
    
ili = intepolate_ili( load_ili('hhs'))
ili_max = ili.max()*13

def calculate_nll(y_true, y_mean, y_std):
    nll = 0.5 * (np.log(2 * np.pi * y_std**2) + ((y_true - y_mean) / y_std)**2)
    return np.mean(nll)
```
