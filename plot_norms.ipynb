```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
region = 'hhs'
ode_name = 'UONN'
num = 1
test_season = 2015
file_prefix = f'norms/{region}/{ode_name}/{test_season}_{num}_'
norm_file = f'{file_prefix}norms.txt'
```

```python
with open(norm_file, 'r') as file:
    file_contents = file.read()

norms = []
for line in file_contents.split('\n'):
    for f in line.split(','):
        try:
            norms.append(float(f))
        except:
            pass

lim = 800

x = np.arange(len(norms))/130
plt.plot(x, norms)
plt.plot([x[0], x[-1]], [lim, lim], color='black')

plt.xlim([3, x[-1]])
plt.ylim([0,1500])
```
